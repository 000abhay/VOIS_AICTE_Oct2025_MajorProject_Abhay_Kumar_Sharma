```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Netflix Dataset.csv")

# Plot Movies vs TV Shows distribution
plt.figure(figsize=(6,4))
df['Category'].value_counts().plot(kind='bar')
plt.title("Distribution of Movies vs TV Shows on Netflix")
plt.xlabel("Category")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Split multiple genres into separate rows
all_genres = df['Type'].dropna().str.split(', ').explode()

# Count top 10 genres
top_genres = all_genres.value_counts().head(10)

# Plot genres
plt.figure(figsize=(8,5))
top_genres.plot(kind='barh')
plt.title("Top 10 Netflix Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
# Count titles by country
top_countries = df['Country'].dropna().value_counts().head(10)

# Plot top countries
plt.figure(figsize=(8,5))
top_countries.plot(kind='barh')
plt.title("Top 10 Countries Contributing to Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```
